```python
%matplotlib inline
import numpy as np
import pylab as pl
import pandas as pd

#Load data
X_df=pd.read_csv('genotypes.csv', delimiter=",")
#X is the matrix of samples and features

y_df=pd.read_csv('phenotype_values.csv', delimiter=",")
#y is the plant biomass (phenotype)

C_df=pd.merge(X_df,y_df,on="sample_ids")
#C.replace("?",100)
C_df
```

```python
#DATA FROM PANDAS TO NUMPY
X=X_df.to_numpy()[:,1:]
y=y_df.to_numpy()[:,1:]
X
```

```python
#SPLITTING DATA BEFORE PREPROCESSING AND IMPUTATION
#we want to separate what we do with the data we have for training and what we have for testing
#for the actual testing for the project we don't need to do this step?
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.4)
#DATA SHAPES
print("X training data shape: ", X_train.shape)
print("X testing data shape: ", X_test.shape)
```

```python
#DATA ENCODING - TRAINING DATA
np.set_printoptions(edgeitems=127)

for j in range(0,X_train[0,:].size):
    values, counts = np.unique(X_train[:,j], return_counts=True)
    values=values[1:]
    counts=counts[1:]
    if(j==30):
        print(X_train[:,j])
    
    X_train[:,j]= np.where(X_train[:,j]==values[np.argmax(counts)],0,X_train[:,j])
    X_train[:,j]= np.where(X_train[:,j]==values[np.argmin(counts)],2,X_train[:,j])
    X_train[:,j]= np.where(X_train[:,j]=="?",np.nan,X_train[:,j])
    if(j==30):
        print(X_train[:,j])
        
#DATA ENCODING - TESTING DATA        
for j in range(0,X_test[0,:].size):
    values, counts = np.unique(X_test[:,j], return_counts=True)
    values=values[1:]
    counts=counts[1:]
    if(j==30):
        print(X_test[:,j])
    
    X_test[:,j]= np.where(X_test[:,j]==values[np.argmax(counts)],0,X_test[:,j])
    X_test[:,j]= np.where(X_test[:,j]==values[np.argmin(counts)],2,X_test[:,j])
    X_test[:,j]= np.where(X_test[:,j]=="?",np.nan,X_test[:,j])
    if(j==30):
        print(X_test[:,j])
```

```python
#DATA CLEANING
#Check if we have at least 70% of missing data ina single sample -> grounds for exclusion of that sample from training dataset
for i in range(0,X_train[:,0].size):
    values, counts = np.unique(X_train[i,:], return_counts=True)
    freq_missing=counts[np.where(np.isnan)]/X_train[0,:].size
    if(freq_missing>=0.7):
        print("Problem at row %i"%i)
#check if we have at least 70% missing data for a single feature of all samples
for i in range(0,X_test[0,:].size):
    values, counts = np.unique(X_test[:,j], return_counts=True)
    freq_missing=counts[np.where(np.isnan)]/X_test[:,0].size
    if(freq_missing>=0.7):
        print("Problem at row %i"%j)
```
